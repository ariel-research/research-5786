## 2. Swap

How do you swap two numbers? Three lines in Java:

```Java
public class Swapper {
    public static void main(String[] args) {
        int a=5, b=10;

        int temp = a; 
        a = b; 
        b = temp
}
```

One line in Python:

````python
a,b = 5,10
print (a,b)

a,b = b,a
print (a,b)
````

You can even swap more than two numbers:

````python
a,b,c = 5,10,15
print (a,b,c)

a,b,c = b,c,a
print (a,b,c)
````
