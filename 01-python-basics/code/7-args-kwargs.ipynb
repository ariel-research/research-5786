## *args

Q: How can we find an average of an unlimited number of variables? 

```python
# option 1: tuple or list: 
def avg_of_tup(my_tup:tuple):
    return sum(my_tup)/len(my_tup)

my_tup = (32,54,82,52)
print(avg_of_tup(my_tup))
print(avg_of_tup((32,54,82,52)))
```

```python

# option 2 : *args
def avg_unlimited(*args):
    return sum(args)/len(args)

print(avg_unlimited(32,54,82,52))

# you can  use '*' to unpack tuple or list:
my_tup = (32,54,82,52)
print(avg_unlimited(*my_tup))
# equivalent to:
print(avg_unlimited(32,54,82,52))
```

```python
# It is easy to send 'args' to another function:

def print_avg(title:str, *args):
    print(f"{title}: {avg_unlimited(*args)}")

print_avg("Average of integers", 1,2,3,4,5,6)
```

## **kwargs

*args    converts __unnamed arguments__ to a __tuple__.

**kwargs converts __named arguments__ to a __dict__.

```python
def print_key_val(**kwargs):
    print(kwargs)
    print(type(kwargs))
    for key, val in kwargs.items():
        print(key, '->', val)

print_key_val(e=1, r=2, j=3, xyz=6)
```

```python
# you can also unpack dict with **kwargs
def f(a, b, c):
    print(f'a = {a}')
    print(f'b = {b}')
    print(f'c = {c}')

d = {'a': 'foo', 'c': 'qux', 'b': 25}
# using the values
f(**d)
# equivalent to
# f(a='foo', b=25, c='qux')
```

```python
# unpack several dictionaries into one dictionary
my_first_dict = {"A": 1, "B": 2, "C":3}
my_second_dict = {"C":4, "D": 5, "E":6}
# for key,val in my_second_dict.items(): my_first_dict[key]=val
my_merged_dict = {**my_first_dict, **my_second_dict}
print(my_merged_dict)
```

```python
def args_kwargs(title, *args, **kwargs):
    print(f'{title}:\nargs={type(args)}={args},\nkwargs={type(kwargs)}={kwargs}')
    print('~~~~~~~~~~~~~~~~~~')

tup = (1,2,3)
dic = {'a':1,'b':2}

args_kwargs("Hello", 1, 2, 3, d=4, e=5, f=6)
```

## Usage in algorithms

```python
def findRoot(f:callable, fPrime:callable, initialGuess:float=0, tolerance:float=0.001):
    x = initialGuess
    while abs(f(x))>tolerance:
        x = x-f(x)/fPrime(x)
    return x
```

```python
def printRoots(**findroot_kwargs:dict):
    print("root of x^2-4 = ", findRoot(lambda x:x**2-4, lambda x:2*x, **findroot_kwargs))
    print("root of x^2-9 = ", findRoot(lambda x:x**2-9, lambda x:2*x, **findroot_kwargs))
    print()

printRoots(initialGuess=1)
printRoots(initialGuess=-1)
printRoots(initialGuess=1, tolerance=0.5)
```

```python
import math
def are_lists_equal(list1, list2, **isclose_kwargs):
    """
    Given two lists, assumed to have the same length, the function checks
    whether the lists are equal.

    The equality check is done via math.isclose.
    **isclose_kwargs are keyword arguments passed to math.isclose.

    >>> are_lists_equal([1,2,3], [1.01,1.99,2.99], rel_tol=0.1, abs_tol=0.1)
    True

    >>> are_lists_equal([1,2,3], [1.01,1.99,2.99], rel_tol=0.001, abs_tol=0.001)
    False
    """
    for a1,a2 in zip(list1,list2):
        if not math.isclose(a1,a2, **isclose_kwargs):
            return False
    return True
```
