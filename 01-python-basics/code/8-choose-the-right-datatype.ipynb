# Data types: choose the right type for the task 

1. `list`:  mutable, order is important.
2. `tuple`: immutable (so can be used as a key), order is important.
3. `set`:   mutable, order is not important (so faster).
4. `frozenset`:   immutable, order is not important.

## list

```python
my_list = [1, 2, 3, '4', 5.0, 6.0, 7.0]
print(my_list)
my_list[3] = "444"
print(my_list)
```

## tuple

```python
my_tuple = (1, 2, 3, '4', 5.0, 6.0, 7.0)
print(my_tuple)
my_tuple[3] = "444" # Error
print(my_tuple)
```

The solution: create a new tuple:

```python
my_new_tuple = my_tuple[0:3] + ("444",) + my_tuple[4:]
print(my_new_tuple)
```

### Only tuples can be used as keys in a dict

```python
my_dict = {
    "a": 1,
    "b": 2,
    ("c","d"): 3
}
print(my_dict)
```

```python
my_dict = {
    "a": 1,
    "b": 2,
    ["c","d"]: 3
}
print(my_dict)
```

### Only tuples can be inserted into a set

```python
my_set = {1, 2, 3, my_tuple}
print(my_set)
```

## frozenset

```python
my_dict = {
    "a": 1,
    "b": 2,
    {"c","d"}: 3
}
print(my_dict)
```

```python
my_dict = {
    "a": 1,
    "b": 2,
    frozenset({"c","d"}): 3
}
print(my_dict)
```

```python
from collections import defaultdict
def count_pairs(mylist:list)->dict:
    counter = defaultdict(int)  # a dict with default value 0.
    for a,b in zip(mylist,mylist[1:]):
        counter[(a,b)] += 1
    return counter
count_pairs([1,2,3,4,5,1,2,1,2,1])
```
