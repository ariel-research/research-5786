## Finding roots

```python
def f(x):
    return (x - 2) * x * (x + 2)**2
x=np.linspace(-3, 3, 1000)
plt.grid()
plt.plot(x, f(x));
```

```python
from scipy.optimize import root
sol = root(lambda x: f(x)-20, x0=2.0)  # find x such that f(x)-20=0. Initial guess: x0
print(sol)
x_sol=sol.x[0]
print(x_sol, f(x_sol))
plt.grid()
plt.plot(x, f(x));
plt.plot(x_sol, f(x_sol), 'x');
```

```python
sol = root(lambda x: f(x)+20, x0=2.0)  # find x such that f(x)+20=0. Initial guess: x0
print(sol)
```

```python
# Find root of two-dimensional function:
sol = root(lambda x: [5*x[0]+7*x[1]-3, 3*x[0]+9*x[1]-5], x0=[0,0])
print(sol)
print(sol.x)
```

## Optimization

```python
from scipy.optimize import minimize_scalar
res = minimize_scalar(f)
plt.grid()
plt.plot(x, f(x));
plt.plot(res.x, res.fun, 'x');
```

```python
print(minimize_scalar(f, bounds=(-3,-1), method='bounded'))
```

```python
from scipy.optimize import minimize

def g(vars):
    x,y = vars
    return (x-1)**2 + (y-2)**2

x=np.linspace(-6, 6, 101)
y=np.linspace(-6, 6, 101)
plt.contourf(x, y, g(np.meshgrid(x,y)))
plt.axis('scaled')
plt.colorbar();
```

```python
minimize(g, (0,0), method='Nelder-Mead')
```

## Linear programming

```python
from scipy.optimize import linprog
# Minimize     6*x + 4*y 
# Subject to   x + y >= 10
#              x  >= 0
#              y  >= 0

# General form:
# Minimize     c*x
# Subject to   A_ub * x <= b_ub

c    = [6, 4]                 # minimization objective
A_ub = [[-1, -1]]             # - x - y <= - 10
b_ub = [-10]                  # 
x_bounds = [(0, None), (0, None)]   # x >= 0, y >= 0

result = linprog(c, A_ub=A_ub, b_ub=b_ub, bounds=x_bounds)

print(f"Optimal value: {result.fun:.2f}")
print(f"Optimal solution: x={result.x[0]:.2f}, y={result.x[1]:.2f}")
```

## Statistics

```python
from scipy.stats import ttest_ind
grades_5784 = np.random.normal(65,10,size=100)
grades_5785 = np.random.normal(60,10,size=100)
plt.hist(grades_5784, bins=100)
plt.hist(grades_5785, bins=100); plt.show()
print(np.average(grades_5784), np.average(grades_5785))
print(ttest_ind(grades_5784, grades_5785))   # Test the hypothesis that the average grade is the same. pvalue = probility that the average is the same.
```
