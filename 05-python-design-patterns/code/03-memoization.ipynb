```python
def fibonacci_recursive(n):
    return  1 if n==1  else  1 if n==2  else  fibonacci_recursive(n-1) + fibonacci_recursive(n-2) 
```

```python
for n in range(1,11):
    print(f'{n}) {fibonacci_recursive(n)}')
fibonacci_recursive(100)
```

```python
fibonacci_cache = {}
def fibonacci_recursive_with_cache(n):
    if n in fibonacci_cache:
        return fibonacci_cache[n]
    value =  1 if n==1  else  1 if n==2  else  fibonacci_recursive_with_cache(n-1) + fibonacci_recursive_with_cache(n-2)
    fibonacci_cache[n] = value 
    return value
```

```python
for n in range(1,101):
    print(f'{n}) {fibonacci_recursive_with_cache(n)}')
```

```python
from functools import lru_cache
@lru_cache(maxsize = 3)
def fibonacci_recursive(n):
    print(n)
    return  1 if n==1  else  1 if n==2  else  fibonacci_recursive(n-1) + fibonacci_recursive(n-2) 
```

```python
for n in range(1,101):
    print(f'{n}) {fibonacci_recursive(n)}')
```
