## Using an iterator

```python
my_list = ['start',' |',' |',' |',' ↓']

my_iterator = iter(my_list)
while True:
    try:
        item = next(my_iterator)
        print(item)
        # if item == " |": break
    except StopIteration:
        print('end')
        break
print("after loop")

# Equivalent to:
for item in my_list:
    print(item)
    # if item == " |": break
else:  # we get here only if we finish looping over all items.
    print('end')
print("after loop")
```

## Defining an iterator

```python
class RangeIterator:
    def __init__(self, first, end, step=1):
        self.state = first
        self.end = end
        self.step = step

    def __iter__(self):
        return self

    def __next__(self):
        if self.state >= self.end:
            raise StopIteration
        res = self.state
        self.state += self.step
        return res

N=1000000
for i in RangeIterator(0,N,2):  # equivalent to: range(0,N,2)
    print(i , end = ' ')
    if i % 5 == 4: 
        break
```

```python
# Fibonacci - example of an iterator for an infinite sequence
class Fibonacci:
    def __init__(self,first,second):
        self.first = first
        self.second = second

    def __iter__(self):
        return self

    def __next__(self):
        res = self.first
        self.first , self.second = self.second , self.first+self.second
        return res

print(f"First elements of fibonacci sequence:")
for res in Fibonacci(0,1):
    print(res ,end = ' ')
    if res>200: break
```

## Defining a generator

```python
def example_of_generator():
    yield "first"
    yield "second"
    yield "third"

for i in example_of_generator():
    print(i)
```

```python
def fibonacci_generator():
    first = 0
    second = 1
    while True:
        yield first
        first , second = second , first+second

print(f"First numbers of fibonacci sequence:")
for fib in fibonacci_generator():
    print(fib,end = ' ')
    if fib>0 and fib%9==0: break
print()
for fib in fibonacci_generator():
    print(fib,end = ' ')
    if fib>0 and fib%9==0: break
```

## [Itertools](https://docs.python.org/3/library/itertools.html)

```python
import itertools
```

```python
print("slice from a string:")
for p in itertools.islice("abcdefgh", 3, 7):
    print(p)
print("slice from a generator:")
for p in itertools.islice(fibonacci_generator(), 3, 7):
    print(p)
```

```python
for p in itertools.permutations("abc"):
    print(p)

for p in itertools.permutations(itertools.islice(fibonacci_generator(),3,7)):
    print(p)
```

```python
for p in itertools.product("abc","def"): 
    print(p)
```

```python
# reminder
for p in zip("abc","def"): 
    print(p)

for p in zip("abc",fibonacci_generator()): 
    print(p)

a = [1,2,3,4,5]
b = fibonacci_generator()
[x+y for x,y in zip(a,b)]
```

```python
for p in itertools.chain(itertools.islice(fibonacci_generator(),3,7), example_of_generator(), itertools.permutations("xyz")):
    print(p)
```

```python
for p in itertools.pairwise(range(5)):
    print(p)
for p in itertools.pairwise(example_of_generator()):
    print(p)
for p in itertools.pairwise([x**2 for x in range(10)]):
    print(p)

# discrete derivative example:
[y-x for x,y in itertools.pairwise([x**2 for x in range(100)])]
```

```python
for p in itertools.count(5):
    print(p)
```

```python
for i in zip(itertools.count(1), ["A","B","C"]):
    print(i)
```
